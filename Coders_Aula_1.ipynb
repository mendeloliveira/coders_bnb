```python
# Atribuição de valores
renda_cliente = 5000
valor_parcela = 1600

# Cálculo do limite (30% da renda)
limite_parcela = renda_cliente * 0.30

# Validação do empréstimo
print(f"Renda: R$ {renda_cliente:.2f}")
print(f"Valor da parcela: R$ {valor_parcela:.2f}")
print(f"Limite permitido (30%): R$ {limite_parcela:.2f}")
print("-" * 30)

if valor_parcela <= limite_parcela:
    print("Empréstimo APROVADO!")
else:
    print("Empréstimo NEGADO: O valor da parcela excede 30% da renda.")
```
